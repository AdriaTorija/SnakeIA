## Representation of the Scores of all Snakes

```python
dfMatrix['x1'] = dfMatrix.index
dfMatrix.plot(kind = 'line',x='x1', y = 'GameScore', title="Scores")
```

```python
df13['x1'] = df13.index
df13.plot(kind = 'line',x='x1', y = 'GameScore', title="Scores")
```

```python
df24['x1'] = df24.index
df24.plot(kind = 'line',x='x1', y = 'GameScore', title="Scores")
```

```python
def getDfScores(df):
    scores=[]
    for i in range(max(df["GameScore"])):
        scores.append(df[df["GameScore"] == i])
    return scores
def drop_duplicates(df):
    return df.drop_duplicates(subset=["GameScore","Steps","StepsDeath","SnakeScore"],keep = 'last').reset_index(drop = True)
```

```python
def plotRepresentation(x):
    i=0
    for p in x:
        stitle="Score: "+str(i)
        plot=p.plot(title=stitle)
        plot.set_xlabel("Snake_Id")
        plot.set_ylabel("Number of Steps")
        i+=1
```

```python
def plotStepsScores(df):
    result=[]
    for i in range(len(df)):
        aux= pd.DataFrame( np.array(df[i]["Steps"]),columns=["Steps"])
        result.append(aux)
    
    return result
```

```python
x=plotStepsScores(getDfScores(drop_duplicates(df24)))
y=plotStepsScores(getDfScores(drop_duplicates(df13)))
z=y=plotStepsScores(getDfScores(drop_duplicates(dfMatrix)))
```

## Representation of Steps depending on the Score

```python
plotRepresentation(z)
```

```python
plotRepresentation(y)
```

```python
plotRepresentation(x)
    
```

## Representation of the best of each generation

```python
def generations(df):
    result = pd.DataFrame(columns=["Best_Score","Generation"])
    gc=0
    generation=0
    maxN=0
    for x in range(len(df)):
        if gc>=100:
            result.loc[len(result)]=(np.array([maxN,generation]))
            
            maxN=0
            gc=0
            generation += 1
        if df["GameScore"][x] > maxN:
            maxN= df["GameScore"][x]
        
        
        gc += 1
    return result
     
```

```python
geneScoreMatrix=generations(dfMatrix)
geneScore13=generations(df13)
geneScore24=generations(df24)
```

```python
geneScoreMatrix.plot(kind = 'line',x='Generation', y = 'Best_Score', title= "GenScores")
```

```python
geneScore13.plot(kind = 'line',x='Generation', y = 'Best_Score', title= "GenScores")
```

```python
geneScore24.plot(kind = 'line',x='Generation', y = 'Best_Score', title= "GenScores")
```
